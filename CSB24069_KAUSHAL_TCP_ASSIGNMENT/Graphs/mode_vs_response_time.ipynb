```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
result=pd.read_csv(r"C:\Users\kaush\Desktop\Internship\CSB24069_KAUSHAL_TCP_ASSIGNMENT\result_table.csv")
message=pd.read_csv(r"C:\Users\kaush\Desktop\Internship\CSB24069_KAUSHAL_TCP_ASSIGNMENT\message_response_log.csv")
```

```python
avg_mode=result.groupby("mode")["average_response_time_seconds"].mean()
```

```python
plt.figure(figsize=(7,5))
avg_mode.plot(kind="bar")
plt.xlabel("Mode")
plt.ylabel("Averagae Response Time(seconds)")
plt.title("Persistent vs new connnection time Response Time")
# plt.savefig("Graphs/mode_vs_response_time.png")
plt.show()
plt.close()
```

```python
avg_throughput=result.groupby("message_size_bytes")["throughput_bytes_per_second"]
```

```python

plt.figure()
avg_throughput.plot(kind="bar")
plt.xlabel("Message Size (bytes)")
plt.ylabel("Throughput (bytes/sec)")
plt.title("Message Size vs Throughput")
# plt.savefig("Graphs/message_size_vs_throughput.png")
plt.show()
plt.close()

```

```python
data_512 = message[message["message_size_bytes"] == 512]

for mode in data_512["mode"].unique():
    data = data_512[data_512["mode"] == mode]

    plt.plot(
        data["message_number"],
        data["response_time_seconds"],
        marker="o",
        label=mode
    )

plt.xlabel("Message Number")
plt.ylabel("Response Time (seconds)")
plt.title("Response Time for 512 Byte Messages")
plt.legend()

# plt.savefig("graphs/message_response_time.png")
plt.show()
plt.close()
```
